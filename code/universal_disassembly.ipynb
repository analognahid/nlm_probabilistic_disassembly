```python
import magic, hashlib, os, traceback
import ntpath
from capstone import *
from capstone.x86 import *
from capstone.arm import *
import collections
import traceback
from elftools.elf.elffile import ELFFile
```

```python
def get_text_section_offset(filename):
    with open(filename, 'rb') as f:
        elffile = ELFFile(f)

        for section in elffile.iter_sections():
            if section.name == '.text':
                return section['sh_offset'] , section['sh_size']

    return None
```

```python
objs_dir_path = '/home/raisul/probabilistic_disassembly/objs'
obj_files = [os.path.join(objs_dir_path, f) for f in os.listdir(objs_dir_path) ]
blob_output_path = '/home/raisul/probabilistic_disassembly/blobs'

for obj_file_path in obj_files:

    
    fh = open(obj_file_path, 'rb')
    bin_bytearray = bytearray(fh.read())
    
    text_section_offset ,text_section_size= get_text_section_offset(obj_file_path)
    print('text_section_offset: ',text_section_offset, 'text_section_size: ', text_section_size)
    bin_bytearray[text_section_offset:text_section_offset+text_section_size]

    with open(obj_file_path, 'rb') as f:
        elffile = ELFFile(f)
        print ('sections:')
        for s in elffile.iter_sections():
            print(s)
        textSection = elffile.get_section_by_name('.text').data()
    
    print(textSection)

    blob_save_path = os.path.join(blob_output_path, os.path.basename(obj_file_path))
    with open(blob_save_path, "wb") as f:
        f.write(textSection)
```
